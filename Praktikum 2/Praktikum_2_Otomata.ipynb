```python
from graphviz import Digraph

class FSMMachine:
    def __init__(self):
        self.start_state = 'S'
        self.accept_state = 'B'

        self.transitions = {
            'S': {'0': 'A', '1': 'B'},
            'A': {'0': 'C', '1': 'B'},
            'B': {'0': 'A', '1': 'B'},
            'C': {'0': 'C', '1': 'C'}
        }

    def check_string(self, input_string):
        state = self.start_state
        path = [state]

        for c in input_string:
            if c not in ['0','1']:
                return None, "Input tidak valid"
            state = self.transitions[state][c]
            path.append(state)

        return state == self.accept_state, path


def draw_fsm(path):
    dot = Digraph(format='png')
    dot.attr(rankdir='LR', splines='true')

    with dot.subgraph() as s:
        s.attr(rank='same')
        s.node('S')

    with dot.subgraph() as s:
        s.attr(rank='same')
        s.node('A')
        s.node('B')

    with dot.subgraph() as s:
        s.attr(rank='same')
        s.node('C')

    dot.node('B', shape='doublecircle')

    dot.node('start', shape='none')
    dot.edge('start', 'S')

    dot.edge('S','A', label='0')
    dot.edge('S','B', label='1')
    dot.edge('A','C', label='0')
    dot.edge('A','B', label='1')
    dot.edge('B','A', label='0')
    dot.edge('B','B', label='1')
    dot.edge('C','C', label='0,1')

    for i in range(len(path)-1):
        u = path[i]
        v = path[i+1]

        dot.edge(
            u, v,
            label=f"{i+1}",
            color='red',
            fontcolor='red',
            penwidth='3',
            constraint='false'
        )

    return dot


def run_fsm(input_string):
    fsm = FSMMachine()
    accepted, path = fsm.check_string(input_string)

    if accepted is None:
        print("ERROR:", path)
        return

    print("Input :", input_string)
    print("Path  :", " -> ".join(path))
    print("Result:", "Accepted" if accepted else "Rejected")

    dot = draw_fsm(path)
    dot.view()

run_fsm("100")
```
